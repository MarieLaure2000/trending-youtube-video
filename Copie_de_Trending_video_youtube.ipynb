```python
!pip install kaggle
import shutil

# Créer le dossier kaggle s'il n'existe pas
!mkdir -p ~/.kaggle

# Déplacer kaggle.json dans le dossier ~/.kaggle/
shutil.move("/content/kaggle.json", "/root/.kaggle/kaggle.json")

# Modifier les permissions du fichier pour éviter les erreurs
!chmod 600 ~/.kaggle/kaggle.json
!kaggle datasets download -d datasnaek/youtube-new
import zipfile

with zipfile.ZipFile("youtube-new.zip", 'r') as zip_ref:
    zip_ref.extractall("/content/youtube_data")
!ls /content/youtube_data
import pandas as pd

# Charger le fichier des vidéos tendances des États-Unis
df = pd.read_csv("/content/youtube_data/USvideos.csv")

# Afficher les 5 premières lignes du tableau
df.head()
# Informations sur le dataset
df.info()

# Statistiques générales des colonnes numériques
df.describe()

# Vérifier les valeurs manquantes
df.isnull().sum()
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.countplot(y=df["channel_title"], order=df["channel_title"].value_counts().index[:10])
plt.title("Top 10 des chaînes avec le plus de vidéos tendances")
plt.xlabel("Nombre de vidéos")
plt.ylabel("Chaînes")
plt.show()
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["views"], y=df["likes"])
plt.title("Relation entre le nombre de vues et le nombre de J'aime")
plt.xlabel("Vues")
plt.ylabel("J'aime")
plt.show()
plt.figure(figsize=(10, 5))
sns.countplot(y=df["category_id"], order=df["category_id"].value_counts().index[:10])
plt.title("Top 10 des catégories les plus populaires")
plt.xlabel("Nombre de vidéos")
plt.ylabel("ID de la catégorie")
plt.show()
```
